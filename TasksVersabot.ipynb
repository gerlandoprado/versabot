```python
def is_weekday_in_range(var):
    
    week = ['DOM', 'SEG', 'TER', 'QUA', 'QUI', 'SEX', 'SAB']

    if "-" in var:
        startday, endday = map(week.index, var.split("-"))
    else:
        startday = endday = week.index(var)

    if startday <= weekday <= endday:
        return True
        
is_weekday_in_range("SEG-QUA")
```

```python
def diff_seconds(time):

    now = datetime.now()
    now = f"{now.hour}:{now.minute}:{now.second}"

    # criar objetos datetime para os horários
    dt1 = datetime.strptime(now, '%H:%M:%S')
    dt2 = datetime.strptime(time, '%H:%M')

    # calcular a diferença em segundos
    diff_seconds = (dt2 - dt1).total_seconds()

    return diff_seconds


print(diff_seconds("11:50"))
```
