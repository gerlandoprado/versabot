```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.service import Service
import datetime
import time
import pandas as pd
import os

versionbot = 1.0

dir_path = os.getcwd()
profile = os.path.join(dir_path, "profile", "wpp")
service = Service('D:\Software\chromedriver.exe')
options = webdriver.ChromeOptions()
options.add_argument(r"user-data-dir={}".format(profile))
nav = webdriver.Chrome(service=service, options=options)

action = ActionChains(nav)
nav.get("https://web.whatsapp.com/")

while len(nav.find_elements(By.ID, "pane-side")) < 1:
    time.sleep(2)
print("Conectado e carregado.")
```

```python
def grupo(grupo):
    
    tentativas = 0
    max_tentativas = 5  # define o número máximo de tentativas
    
    while tentativas < max_tentativas:
        try:
            nav.find_element(By.XPATH, "//div[@title='Caixa de texto de pesquisa']").send_keys(grupo)
            time.sleep(0.5)
            nav.find_element(By.XPATH, f"//span[@title='{grupo}']").click()
            nav.find_element(By.XPATH, "//div[@title='Caixa de texto de pesquisa']").click()
            action.key_down(Keys.CONTROL).send_keys('a').key_up(Keys.CONTROL).perform()
            action.key_down(Keys.DELETE).key_up(Keys.DELETE).perform()
        except:
            print("Erro: " + grupo)
            action.key_down(Keys.CONTROL).send_keys('a').key_up(Keys.CONTROL).perform()
            action.key_down(Keys.DELETE).key_up(Keys.DELETE).perform()
            tentativas += 1  # incrementa o número de tentativas
        else:
            print("Entrou: " + grupo)
            break  # se a função executar sem erros, sai do loop
    
    if tentativas == max_tentativas:
        print("Número máximo de tentativas atingido.")
```

```python
def altgrupo(nome, status):
    grupo(nome)
    time.sleep(0.5)
    nav.find_element(By.XPATH, "/html/body/div[1]/div/div/div[5]/div/header/div[2]/div[1]/div/span").click()
    time.sleep(0.5)
    nav.find_element(By.XPATH, "//*[contains(text(),'Configurações do grupo')]").click()
    time.sleep(0.5)
    nav.find_element(By.XPATH, "//*[contains(text(),'Enviar mensagens')]").click()
    time.sleep(0.5)
    nav.find_element(By.XPATH, f"/html/body/div[1]/div/span[2]/div/div/div/div/div/div/div[2]/form/ol/li[{status}]/label").click()
    nav.find_element(By.XPATH, "//*[contains(text(),'Confirmar')]").click()
                                
    
    time.sleep(1)
    action.key_down(Keys.ESCAPE).key_up(Keys.ESCAPE).perform()
    action.key_down(Keys.ESCAPE).key_up(Keys.ESCAPE).perform()
```

```python
def verifica_horario():
    while True:
        agora = datetime.datetime.now()
        print(agora)
        if agora.hour == 22 and agora.minute == 0 and agora.second >= 0 and agora.second < 5:
            print("Já são 22h!")
            break
        time.sleep(5)  # aguarda 5 segundos
```

```python
class mensagem:
    
    def __init__(self, num):
        
        sel = nav.find_elements(By.CLASS_NAME, "_3nbHh")
        
        try:
            self.text = sel[num].find_element(By.CLASS_NAME, "_21Ahp").text
            self.hour = sel[num].find_element(By.CLASS_NAME, "fewfhwl7").text
        except:
            self.text = "null"
            self.hour = "null"
        
        try:
            self.display = sel[num].find_element(By.CLASS_NAME, "_2ne0X").text
            
            cell = sel[num].find_element(By.CLASS_NAME, "WJuYU").text
            cellformat = ''.join(filter(str.isdigit, cell))
            self.cell = cellformat
            
        except:
            self.display = "null"
            self.cell = "null"

    def getall():
        sel = nav.find_elements(By.CLASS_NAME, "_3nbHh")
        mensagens = []
        for i in range(len(sel)):
            mensagens.append(mensagem(i))
        return mensagens
```

```python
def enviar(mensagem):
    
    xpath = nav.find_element(By.XPATH, "/html/body/div[1]/div/div/div[5]/div/footer/div[1]/div/span[2]/div/div[2]/div[1]/div/div[1]/p")
    
    if isinstance(mensagem, list):
        
        for linha in mensagem:
            xpath.send_keys(linha)
            action.key_down(Keys.SHIFT).send_keys(Keys.ENTER).key_up(Keys.SHIFT).perform()
            
    else:
        xpath.send_keys(mensagem)
        
    action.key_down(Keys.ENTER).key_up(Keys.ENTER).perform()  
```
